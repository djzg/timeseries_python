## Time-series data

A sequence of information which attaches a time period to each value

To be able to analyze time-periods, they all have to be equal and clearly defined which would result in constant frequency

### Notation in time-series

Capital letters used for entire period, lowercase used for a single period

T = entire year

t = single day

Xt Xt-1 Xt+1

### Peculiarities in time-series data

Intervals need to be identical

Graphs do not follow standard distributions

### QQ Plot

```python
# Quantile-quantile plot 
# Used to determine whether a data set is distributed a certain way
# Usually showcases how the data fits a Normal Distribution
```

Takes all the values a variable can take, and arranges them in accending order

Theoretical quantiles - how many standard deviations away from the mean these values are

Red line shows what the data points should follow, if they are Normally Distributed

```python
scipy.stats.probplot(df_comp.spx, plot=pylab)
pylab.show()
```
